```python
#creating a copy of df_joined

df_copy= df_joined.copy(deep=False)
```

```python
df_copy["Date"]=pd.to_datetime(df_copy["Date"])
```

```python
df_copy = df_copy.sort_values("Date")
```

```python
df_copy.head(5)
```

```python
#Monthly Sales Trend

monthly_sales = (
    df_copy.set_index("Date")
    .resample("ME")["Weekly_Sales"]
    .sum()
)
monthly_sales.head()
```

```python

plt.figure(figsize=(12,6))
monthly_sales.plot()
font1={"font":"serif", "color":"blue", "size":20}
font2={"font":"serif", "color":"darkred", "size":15}
plt.title("Montly Sales Trend", fontdict=font1)
plt.xlabel("Month", fontdict=font2)
plt.ylabel("Average Sales", fontdict=font2)
plt.grid(axis="y")
```

```python

monthly_sales_ma = monthly_sales.rolling(window=3).mean()

plt.figure(figsize=(12,6))
plt.plot(monthly_sales, label="Actual Sales")
plt.plot(monthly_sales_ma, label="3-Month MA", linewidth=3)
plt.legend()
plt.title("Sales Trend with Moving Average")
plt.show()
```

```python
#Revenue by region
monthly_store = (
    df_copy
    .groupby(["Store", pd.Grouper(key="Date", freq="ME")])
    ["Weekly_Sales"]
    .sum()
    .reset_index()
)
```

```python
pivot_store = monthly_store.pivot(
    index="Date",
    columns="Store",
    values="Weekly_Sales"
)

pivot_store.plot(figsize=(12,6))
plt.title("Store Sales Over Time")
plt.show()
```
